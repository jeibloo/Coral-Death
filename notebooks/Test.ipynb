```python
import plotly.express as px
import pandas as pd

coral = pd.read_csv("./dataset/CoralBleachingClean.csv",index_col='ID')
```

```python
coral.drop("Unnamed: 0",axis=1,inplace=True)
coral.head()
```

```python
px.scatter_geo(coral,
            lat='LAT',lon='LON',
            color="BLEACHING_SEVERITY",
            projection="orthographic",hover_name='LOCATION',
            animation_frame='YEAR', range_color=(0,1),
            width=None,height=None)
```
